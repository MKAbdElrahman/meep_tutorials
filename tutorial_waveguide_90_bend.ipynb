```python
import meep as mp 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
cell_size  = mp.Vector3(16,16)
```

```python
resolution = 10
```

```python
pml =  [mp.PML(1.0)]
```

```python
M = mp.Medium(epsilon = 12.0)
```

```python
horz_wg = mp.Block(mp.Vector3(10,1),material = M, center = mp.Vector3(-5,0))
vert_wg = mp.Block(mp.Vector3(1,10),material = M, center = mp.Vector3(0,-4.5))
geometry = [horz_wg, vert_wg]
```

```python
sources = [mp.Source(mp.ContinuousSource(wavelength = 2 * np.sqrt(12)),
                     component = mp.Ez,
                     center = mp.Vector3(-7.0,0,0),
                     size = mp.Vector3(0,1)
                    )]
```

```python
sim = mp.Simulation(cell_size,resolution, geometry = geometry,boundary_layers = pml,sources = sources)
sim.plot2D()
```

```python
sim.run(until = 1000)
```

```python
plt.imshow(sim.get_array(component = mp.Ez).transpose(),cmap = "RdBu")
```
